```python
import pandas as pd
import nltk
from nltk.tokenize import word_tokenize
```

```python
df = pd.read_csv("job_data_files/seniority_labelled_development_set.csv")

# Get unique values from the 'y_true' column
unique_values = df["y_true"].unique().tolist()

# Print the results
print("Unique y_true values:", unique_values)
print("Number of unique y_true values:", len(unique_values))
```

```python
def rule_based_seniority(row):
    # Combine relevant fields into a single text string
    text = ' '.join([
        str(row['job_title']), 
        str(row['job_summary']), 
        str(row['job_ad_details'])
    ]).lower()

    # Tokenize the text
    tokens = word_tokenize(text)

    for v in unique_values:
        if v in tokens:
            return v
    return "entry level"
```

```python
df = pd.read_csv("job_data_files/seniority_labelled_test_set.csv")

# Apply the rule-based function to each row
df['predicted'] = df.apply(rule_based_seniority, axis=1)

# Calculate the accuracy rate by comparing y_true and predicted
accuracy = (df['y_true'] == df['predicted']).mean()

print("Accuracy rate:", accuracy)
```
